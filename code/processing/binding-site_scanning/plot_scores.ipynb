```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt

import regseq2

regseq2.viz.matplotlib_style()
```

```python
df = pd.read_csv("../../../data/binding-site_scanning/yadI_predicted_-45_-20.tsv", sep='\t', comment='#')
df['TF'] = [x.split('_')[0] for x in df['Target_ID']]

fig, ax = plt.subplots(figsize=(1.5, 2))
ax.set_xticks([])
x = np.random.randn(len(df))
inds = df['TF'] == 'CRP'
inds_not = df['TF'] != 'CRP'
ax.scatter(x[inds], df['q-value'].values[inds], s=10, color="#eac264")
ax.scatter(x[inds_not], df['q-value'].values[inds_not], s=10)
ax.set_yscale('log')
ax.set_ylabel('q-value')

plt.tight_layout()
fig.savefig('../../../figures/yadI_comp.pdf')
```
